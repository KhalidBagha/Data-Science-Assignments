```python
meanVal  = data["Speed"].mean()
def setSpeedStatus(x):
    if x > meanVal:
        return "High Speed"
    elif x == meanVal :
         return "Equal Speed"
    else:
        return "Low Speed"
data["High_Low_Speed"]=data["Speed"].apply(setSpeedStatus)

data
```

```python
meanVal  = data["HP"].mean()
print(meanVal)
def setHPStatus(x):
    if x > meanVal:
        return "High HP"
    elif x == meanVal :
         return "Equal HP"
    else:
        return "Low HP"
data["High_Low_HP"]=data["HP"].apply(setHPStatus)

data
```
